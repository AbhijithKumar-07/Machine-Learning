```python
# Range of elements in the Numpy Array -> Range = [startPoint, endPoint)
# np.linspace(startPoint, endPoint, No.of Elements)
# np.linspace() -> Creates a Numpy array with given No.of elements required
# np.linspace() -> Difference between consecutive elements in the Numpy Array is Same
d1 = np.linspace(10, 30, 5)
print(d1)
```

```python
# In the above example the difference between the consecutive elements in the Numpy array is Same
# 15-10 = 5
# 20-15 = 5
# 25-20 = 5
# 30-25 = 5
```

```python
# Creating a Numpy Array By providing a 'Step Value' to the elements of the array
# np.arange(startPoint, endPoint, stepValue)
d2 = np.arange(10, 30, 5)
print(d2)
```



> *Mathematical Operations On a Numpy Array*



```python
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]

print(list1 + list2)    # Concatenate or Joins two Lists
```

```python
a = np.random.randint(0, 10, (3, 3))
b = np.random.randint(10, 20, (3, 3))
```

```python
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)

# Here the above Mathematical Operations Is Performed between the Corressponding Elements of the 2 Matrices (Numpy Arrays)
# For a*b => Matrix Multiplication is Not performed => a*b => Multiplication is performed between the Corressponding Elements Of the 2 Numpy Arrays
```

```python
a = np.random.randint(0, 10, (3, 3))
b = np.random.randint(10, 20, (3, 3))

print(a)
print(b)
```

```python
print(np.add(a, b))         # a+b
print(np.subtract(a, b))    # a-b
print(np.multiply(a, b))    # a*b
print(np.divide(a, b))      # a/b
```

```python
# np.multiply(a, b)  => Here Matrix Multiplication will not be takes place
# np.multiply() => Multiplication is performed between Corressponding Elements of the 2 Numpy Arrays
```



> ***Numpy Array => Matrix***



```python
A = np.random.randint(0, 10, (2, 3))
print(A)
print(A.shape)
```



> *Transpose Of a Numpy Array*
- Method 1 ->  *np.transpose(np_array)*
- Method 2 ->  *np_array.T*




```python
# Transpose of a Numpy Array (Matrix)
A_transpose = np.transpose(A)
print(A_transpose)
print(A_transpose.shape)
```

```python
# Transpose of a Numpy Array
A_transpose1 = A.T
print(A_transpose1)
print(A_transpose1.shape)
```
