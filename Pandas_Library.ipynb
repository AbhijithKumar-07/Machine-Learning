

> ***Manipulating a Data Frame***



```python
from sklearn.datasets import load_diabetes
```

```python
diabetes_dataset = load_diabetes()

print(diabetes_dataset)
```

```python
diabetes_dataframe = pd.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
```

```python
diabetes_dataframe.head()
```

```python
# Adding a Column to a Data Frame
diabetes_dataframe["newColumn"] = diabetes_dataset.target
```

```python
diabetes_dataframe.head()
```

```python
# Adding a Column to a Data Frame
# df["colName"] = [ele1, ele2, ele3, .......]
# Here total No.of elements should be Equal to the No.of Rows in the Data Frame
```

```python
# Removing a Row from a Data Frame
diabetes_dataframe.drop(index=0, axis=0)
```

```python
# Removing a Column from a Data Frame
diabetes_dataframe.drop(columns="s1", axis=1)
```

```python
# Accessing a Particular Row in a Data Frame using the Index value
diabetes_dataframe.iloc[2]
```

```python
# Accessing a Particular Column in a Data Frame
# df.iloc[ : , colIndex ]
diabetes_dataframe.iloc[:, 0]
```

```python
print(diabetes_dataframe.iloc[:, 0])   # First Column
print(diabetes_dataframe.iloc[:, 1])   # Second Column
print(diabetes_dataframe.iloc[:, -1])  # Last Column
```



> ***Correlation***
- ***Positive Correlation*** - *If two Columns are Directly Proportional*
- ***Negative Correlation*** - *If two Columns are Inversely Proportional*



```python
diabetes_dataframe.corr()
```
