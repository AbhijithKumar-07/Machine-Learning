# Operators

1. Arithmetic Operators

```python
num_1 = 2
num_2 = 4

# addition
sum = num_1 + num_2

# Subraction
sub = num_1 - num_2

# Multiplication
product = num_1 * num_2

# Division
quotient = num_1 / num_2

# Modulus Operator -> Remainder
remainder = num_1 % num_2

# Exponentiation
power = num_1 ** num_2      #  num_1 ^ num_2 = 2 ^ 4 = 16

print("Sum: ", sum)
print("Sub: ", sub)
print("Product: ", product)
print("Quotient: ", quotient)
print("Remainder: ", remainder)
print("Power: ", power)
```

```python
a = 7
b = 11
c = 45

if (a>b):
  if(a>c):
    print("a is the greatest number")
  else:
    print("c is the greatest number")
else:
  if(b>c):
    print("b is the greatest number")
  else:
    print("c is the greatest number")

```

# Functions

```python
def factorial(num):
  if num == 0:
    return 1
  else:
    fact = 1
    for i in range(1, num+1):
      fact = fact * i
    return fact

print(factorial(5))
```

```python
print(factorial(10))
```

```python
print(factorial(4))
```
